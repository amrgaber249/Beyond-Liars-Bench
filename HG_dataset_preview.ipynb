```python
import os
from datasets import load_dataset, get_dataset_config_names
from huggingface_hub import login

# get HF token from colab secrets
hf_token = os.environ.get("HF_TOKEN")
if hf_token:
    login(token=hf_token)
else:
    print("HF_TOKEN not found in environment; trying public access")

DATASET_ID = "Cadenza-Labs/liars-bench"

# dataset configs
print("Listing dataset configs for:", DATASET_ID)
configs = get_dataset_config_names(DATASET_ID)
print("Configs found:", configs)

# loading each configs
for cfg in configs:
    print("\n" + "="*80)
    print(f"CONFIG: {cfg}")
    try:
        ds = load_dataset(DATASET_ID, cfg, use_auth_token=hf_token)
    except Exception as e:
        print("Failed to load config:", cfg)
        print("Error:", e)
        continue

    print("Splits:", list(ds.keys()))

    for split_name, split_ds in ds.items():
        print("\n--- Split:", split_name)
        print("Number of rows:", len(split_ds))
        print("Columns:", split_ds.column_names)

        # print a couple of rows
        n_show = min(3, len(split_ds))
        for i in range(n_show):
            print(f"\nExample {i}:")
            print(split_ds[i])
```
